Lets fixed $\Lambda = 4$

```julia
using IntervalArithmetic
```

```julia
Λ = Interval(4)
```

```julia
α = Λ/(Λ-1)
```

```julia
K = Int64(floor(mid(Λ)))
```

```julia
q_K_k(k) = max(0, 0.5*((Λ-1-k)/(Λ-1)))
```

```julia
Finding the branch intervals. the branch intervals.F(k) = x -> x + α*(k-(k^2-(k/α)*(k+1-2*x))^(0.5))
```

Finding the branch intervals.

```julia
[q_K_k(i) for i in 1:K]
```

```julia
using Plots
```

```julia
f1 = F(1)
plot(x->mid(f1(x),1), (1/3), 1)
```

```julia
f2 = F(20.5)
plot(x->mid(f2(x),1), (1/6), (1/3))
```

```julia
f3 = F(3)
plot(x->mid(f3(x),1), 0, (1/6))
```

```julia
using RigorousInvariantMeasures
```

```julia
D = PwMap([x->f3(x)-3.0;x->f2(x)-2.0; x->f1(x)-1.0], [Interval(0), Interval(1/6),Interval(1/3), Interval(1)])
```

```julia
B = Ulam(1024)
```

```julia
Q = DiscretizedOperator(B, D)
```
