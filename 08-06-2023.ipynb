# Gora and Boyarsky Rotary Drill.

The interest in the problem of Drill maps arises from the fact that the motion of an oil drill gives rise in a natural way to a discontinuous, piecewise expanding transformation on an interval into itself for sufficiently large rotational speeds.

- The behavior of the drill depends critically on the number $\Lambda = \frac{v^2 M}{FR}$, known as Froude's number. When we have $\Lambda < 1$ the trajectory of center of the Drill is the base curve $p(x) = \sqrt{(R^2 - (x-a_k)^2)}$ where for $kT < x < (k + 1)T$, $k = 0, 1,2, ...$ ,. Intuitively, this is the case when the cylinder is rotating sufficiently slowly so that at least one tooth is always in contact with the x-axis.



- We are interested in study the case $\Lambda > 1$. This means that the cylinder is being pulled away from the x-axis and that the center of the cylinder moves above the base curve. This information reflects the fact
that the drill jumps away (from the x-axis) and then hits the rock as it goes down.

Using Newton's Law, we have the following differential equation for

$$

y(x) = M  \frac{d^2y}{dt^2} = -F

$$

or, since $x = vt$

$$

\frac{d^2y}{dx^2} = - \frac{F}{Mv^2}

$$

Let $x_0$ be the first point where the trajectory $y(x)$ of $C$ goes above the base curve $p(x)$. The idea here is find the point sequence of intersection in between $y(x)$ and $p(x)$.

1. We need make the variables change, where $\bar{x} = \frac{x}{T}$ and $u = \frac{2R}{T^2}y + \left( \frac{1}{4} - \frac{2R^2}{T^2} \right)$ and so we have to write our EDO and base curve this terms.


After of some calculations returning now to our main consideration, we have
$$

\bar{x}_{n+1} = \bar{x}_n + d_{\Lambda}(\bar{x}_n),

$$
where 
$$

d_{\Lambda}(\bar{x}) = α(k-\sqrt{k^2-\left(\frac{k}{α}\right)(k+1-2\bar{x})})

$$

- $q_{K-k} < \bar{x} \le q_{K-k+1}$

- $k = 1, 2, ..., [\Lambda]$

The sequence of "hitting points", ${\hat{\bar{x}}_n}$ where $\hat{\bar{x}}_n = \bar{x_n} - [\bar{x}_n]$ can be represented as a trajectory of a transformation:
$$

\tau_{\Lambda}(\bar{x}) = \bar{x} + d_{\Lambda}(\bar{x})

$$

The behavior of the drill depends on properties of this transformation. $\tau_{\Lambda}(\bar{x})$ is piecewise
monotonic and its properties depend on $\Lambda$.

Not we consider the case where $1 \le \Lambda < 2$, because this case $\tau_{\Lambda}(\bar{x})$ has an an attracting fixed point.

For $\Lambda > 2$ and $k = 2,3, ... ,K$, $\sqrt{\Lambda k } > 2$. Therefore, the derivative of $\tau_{\Lambda}$ can have modulus less than 1 only if $k = 1$.


#### Hence for $\Lambda > 2$, we have


# 1. $\left| \frac{d\tau_{\Lambda}}{d\bar{x}} \right| > 1$ on  $\left[ 0,1 - \frac{3\Lambda}{8(\Lambda - 1)} \right]$


# 2. $\left| \frac{d\tau_{\Lambda}}{d\bar{x}} \right| > 1$ on  $\left[1-\frac{3\Lambda}{8(\Lambda - 1)}, 1 \right]$


# 3. $\frac{d\tau_{\Lambda}}{d\bar{x}}(1)  =0$

The transformation is not expanding on $[0,1]$, but we can still extract a piecewise expanding transformation in one of two ways:
- by restricting $\tau_{\Lambda}$ to an interval $[0,1 - \varepsilon]$ that is invariant, or

- by considering some iterate $\tau_{\Lambda}^n$ that is expanding.

# Attention 

- Since $\tau_{\Lambda}(0) = 1$ for $\Lambda = 3,4,5,...$, we cannot have an invariant set
of the form $[0, 1-\varepsilon]$ for these values of $\Lambda$.

- If us assume that $\Lambda > 2$ and $\Lambda \not\in \{3,4,5,...\}$ and let $\bar{k} = [\bar{x}]-1$. Furthemore the derivative of $\tau_{\Lambda}(\bar{x})\mid_{[0, 1-\varepsilon]}$ is not zero.
